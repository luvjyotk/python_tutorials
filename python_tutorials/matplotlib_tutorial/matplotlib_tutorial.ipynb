### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# you can also pass hex codes as colors
# plt.plot(x,y, label = '2x', color = '#53cba4', linewidth = 3, linestyle='--', marker = ".", markersize=10, markeredgecolor='#734763')

# we can also use shorthand notaion
# fmt = '[color][marker][line]'
plt.plot(x,y,'b.--', label='2x')

## line number 2
x2 = np.arange(0,4,0.5)
plt.plot(x2,x2**2,'r',label='x^2')

plt.title('Our First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname': 'Cambria'})
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])  # x ticks
plt.yticks([0,2,4,6,8,7,10])

# adding legends
## for adding a legend, we need to specify a label in the plot function
plt.legend()

plt.show()
```
