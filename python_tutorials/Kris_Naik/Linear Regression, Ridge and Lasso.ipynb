# Linear Regression, Ridge and Lasso

```python
# house price dataset
from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = load_boston()
```

```python
type(df)
```

```python
pd.DataFrame(df.data)
```

```python
dataset = pd.DataFrame(df.data)
dataset.columns=df.feature_names #feature_names is column in dataset
print(dataset.head())
```

```python
# making my dependent variable
dataset['Price'] = df.target
```

```python
dataset.head()
```

```python
# dividing the dataset into independent and dependent features
X = dataset.iloc[:,:-1] ## independent feature # just skipping the last column
y = dataset.iloc[:,-1] ## dependent features
```

```python
X.head() # shows me all the dependent features
```

```python
y.head() # shows me all the indep features
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# cross_val_score --> lots of train and test splits
lin_reg = LinearRegression()

# applying cross validation: divide train and test data in such a way that every combination of train and test data is taken 
# by the model and whosever accuracy is better, that entire thing is combined

mse = cross_val_score(lin_reg,X,y, scoring='neg_mean_squared_error',cv=5)
print(mse)

mean_mse = np.mean(mse)
print(mean_mse)
```

```python
# to do predition

lin_reg.predict()
```
